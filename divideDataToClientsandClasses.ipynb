```python
import numpy as np
import os
import glob
import PIL
import PIL.Image
import tensorflow as tf
import tensorflow_datasets as tfds
```

```python
dsdir = "C:/Users/temmuz/Desktop/2022-2/proje/stanford/dataset"
for i in range(6):
    dsdiri = dsdir+"/"+str(i)
    if not os.path.exists(dsdiri):
        os.mkdir(dsdiri)
    for j in range(11):
        dsdirij = dsdiri+"/"+str(j)
        if not os.path.exists(dsdirij):
            os.mkdir(dsdirij)
```

```python
places = ["office","conferenceRoom","auditorium","lobby","lounge","hallway","copyRoom","pantry","openspace","storage","WC"]
```

```python
data_dirs = ["E:/scene/noXYZ_area_1_no_xyz/rgb","E:/scene/noXYZ_area_2_no_xyz/rgb","E:/scene/noXYZ_area_3_no_xyz/rgb",
            "E:/scene/noXYZ_area_4_no_xyz/rgb","E:/scene/noXYZ_area_5a_no_xyz/rgb","E:/scene/noXYZ_area_5b_no_xyz/rgb",
            "E:/scene/noXYZ_area_6_no_xyz/rgb"]
```

```python
for dtdirc,data_dir in enumerate(data_dirs):
    image_names = glob.glob(data_dir+'/*.png')
    loccount = np.zeros(11,dtype=int)
    print(dtdirc,"len: ",len(image_names))
    for imc,name in enumerate(image_names):
        if imc%100==0:
            print(imc)
        im1=PIL.Image.open(name)
        im2 = im1.resize((112,112))
        for j,place in enumerate(places):
            if place in name:
                savedir = dsdir+"/"+str(dtdirc)+"/"+str(j)+"/"+str(loccount[j])+".jpg"
                im2.save(savedir)
                loccount[j] += 1
                break
```
